```python
import matplotlib.pyplot as plt

x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]

plt.figure(figsize=(8, 6))

plt.plot(x, y, linestyle='--', color='red', linewidth=3, label='Ligne')

plt.scatter(x, y, color='blue', s=12**2, label='Points')  # Taille des points

plt.xlim(1, 8)
plt.ylim(1, 8)

plt.xlabel("Axe des X")
plt.ylabel("Axe des Y")
plt.title("Graphique avec des Pics")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]

fig, ax1 = plt.subplots()

ax1.plot(x_axis, left_data, color='black', label='Big')
ax1.set_xlabel('Axe X')
ax1.set_ylabel('Valeurs Gauche', color='black')
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()

ax2.plot(x_axis, right_data, color='red', label='Small')
ax2.set_ylabel('Valeurs Droite', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Twin Axes")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.5)  

for i, ax in enumerate(axes.flat, start=1):  
    ax.set_title(f"Titre {i}")  
    ax.text(0.5, 0.5, f"(2,3,{i})", ha='center', va='center', fontsize=12)  
    
    ax.set_xticks(np.arange(0, 1.1, 0.5))  
    ax.set_yticks(np.arange(0, 1.1, 0.25)) 
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    ax.set_xticks(ax.get_xticks(), minor=True)
    ax.set_yticks(ax.get_yticks(), minor=True)

plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

returns = np.random.randn(50)
price = 100 + np.cumsum(returns)
dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price), columns=['Date', 'Company_A'])

fig = px.line(df, x='Date', y='Company_A', title="Company A's Stock Price")

fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Company A's Price"
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

returns = np.random.randn(50)
price = 100 + np.cumsum(returns)
dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price), columns=['Date', 'Company_A'])

fig = go.Figure()

fig.add_trace(go.Scatter(x=df['Date'], y=df['Company_A'], mode='lines', name='Company_A'))

fig.update_layout(
    title="Company A's Stock Price",
    xaxis_title="Date",
    yaxis_title="Company A's Price"
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

y0 = np.random.randn(50)
y1 = np.random.randn(50) + 1  
y2 = np.random.randn(50) + 2  

fig = go.Figure()

fig.add_trace(go.Box(y=y0, name='Distribution 0'))
fig.add_trace(go.Box(y=y1, name='Distribution 1'))
fig.add_trace(go.Box(y=y2, name='Distribution 2'))

fig.update_layout(
    title="Comparaison des distributions",
    xaxis_title="Distributions",
    yaxis_title="Valeurs",
    showlegend=True
)

fig.show()
```
